## Odoo: CREATE, Monday: READ

```python
import random

prompt = "From the Monday board whose 'board_id' is given in the input, fetch all the applicants whose 'name' is provided in the input and add them to the list of job applicants in Odoo."

dataset = []
for _ in range(1000):
    board_id = random.randint(4000000000, 6000000000)
    new_prompt = prompt.format(board_id)
    with open('dataset_generation_use_case_1.py', 'r') as py_file:
        file_contents = py_file.read()
        example = (new_prompt, file_contents)
        dataset.append(example)

with open('dataset_use_case_1_examples.txt', 'w') as file:
    for example in dataset:
        (prompt, code) = example
        file.writelines(['-----EXAMPLE-----\n', '-----PROMPT-----\n', prompt + '\n', '-----CODE-----\n', code + '\n'])
```

## Odoo: UPDATE, Monday: READ

```python
import random

prompt = "From the Monday board whose 'board_id' is given in the input, read the names and statuses on Monday and update them on Odoo. The Monday field ids of names and status are '{}' and '{}' respectively. The Odoo field ids of names and statuses are 'partner_name' and 'stage_id' respectively. Use the following mapping to convert from statuses on Monday to stage ids on Odoo: '{}'->1,'{}'->2,'{}'->3,'{}'->4,'{}'->5,'{}'->6."
monday_name_ids = ['name', 'applicant_name', 'full_name', 'item_name']
monday_status_ids = ['text', 'status0', 'status1', 'status2', 'status3', 'status4', 'status5', 'status6', 'status7', 'status8', 'status9', 'status10', 'status11', 'status12', 'status_id', 'text2', 'text3', 'text4', 'text5', 'text6', 'text8', 'text9', 'text10']
monday_statuses_list = [['New', 'Initial Qualification', 'First Interview', 'Second Interview', 'Contract Proposal', 'Contract Signed'], ['New', 'First Interview', 'Second Interview', 'Third Interview', 'Contract Proposal', 'Contract Signed'], ['HR screening', 'Phone screening', 'First Interview', 'Coding round', 'System Design', 'Final Interview'], ['New', 'Initial Qualification', 'First Interview', 'Second Interview', 'Third Interview', 'Decision Interview'], ['First', 'Second', 'Group Interview', 'HR round', 'Contract Proposal', 'Contract Signed'], ['Phone screen', 'Initial Qualification', 'First Interview', 'Second Interview', 'Contract Proposal', 'Offer signed']]
dataset = []
for _ in range(10000):
    monday_name_id = random.choice(monday_name_ids)
    monday_status_id = random.choice(monday_status_ids)
    monday_statuses = random.choice(monday_statuses_list)
    first = monday_statuses[0]
    second = monday_statuses[1]
    third = monday_statuses[2]
    fourth = monday_statuses[3]
    fifth = monday_statuses[4]
    sixth = monday_statuses[5]
    new_prompt = prompt.format(monday_name_id, monday_status_id, first, second, third, fourth, fifth, sixth)
    with open('dataset_generation_use_case_2.py', 'r') as py_file:
        file_contents = py_file.read()
        example = (new_prompt, file_contents)
        dataset.append(example)

with open('dataset_use_case_2_examples.txt', 'w') as file:
    for example in dataset:
        (prompt, code) = example
        file.writelines(['-----EXAMPLE-----\n', '-----PROMPT-----\n', prompt + '\n', '-----CODE-----\n', code + '\n'])
```
